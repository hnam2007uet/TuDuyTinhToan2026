## W5A3
#### Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def is_prime (n):
  if n<2:
    return False
  for i in range (2, int (n**0.5) + 1):
    if n%i == 0:
      return False
  return True
n = int (input ())
is_prime (n)
```

## W5A4
#### Viết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def check_num (n):
  summ = 1
  for i in range (2, int (n**0.5) + 1):
    if n%i==0:
      summ+=i
      summ+=n//i
  if summ == n:
    return True
  return False
n = int (input ())
check_num (n)
```

#### Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
#### Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
#### Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn.

```python
def nhiphan (n):
  tmp = ""
  while n!=0:
    tmp = str (n%2) + tmp
    n//=2
  return tmp
def _dis (a, b):
  s1 = nhiphan (a)
  s2 = nhiphan (b)
  kq = 0
  dis = abs (len (s1) - len (s2))
  if len (s1) > len (s2):
    for i in range (dis):
      s2 = "0" + s2
  else:
    for i in range (dis):
      s1 = "0" + s1
  for i in range (len (s1)):
    if s1[i]!=s2[i]:
      kq+=1
  return kq
a, b = map (int, input ().split ())
_dis (a,b)

```

Thứ tự của các kí tự được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'. Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai từ đầu vào có cùng độ dài.


```python
def dangcau (a, b):
  dem1 = {}
  dem2 = {}
  for i in range (len (a)):
    c1 = a[i]
    c2 = b[i]
    if c1 in dem1:
      if dem1[c1] != c2:
        return False
    else:
      dem1[c1] = c2
    if c2 in dem2:
      if dem2 [c2] !=c1:
        return False
    else:
        dem2[c2] = c1
  return True
a, b = map (str, input ().split ())
dangcau (a,b)
```

## W5A12
#### Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|.

```python
def _divide (sbc, sc):
  if sc == 0:
    return -1
  q = sbc//sc
  r = sbc % sc
  return q, r
a, b = map (int, input ().split ())
q, r = _divide (a,b)
print (q, r)
```

## W5A15
#### Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
def isrighttriangle (a,b,c):
  if a+ b <= c or a+c<=b or b+c <= a:
    return False
  if a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
    return True
  return False
a,b,c = map (int, input ().split ())
isrighttriangle (a,b,c)
```
