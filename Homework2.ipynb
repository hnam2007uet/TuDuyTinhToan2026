## Bài 3
#### Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, vuông cân, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình

```python
a, b, c = map (int, input ().split ())
if a+b<=c or a+c<=b or b+c<=a:
  print ("Không phải tam giác")
elif a==b==c:
  print ("Tam giác đều")
elif a==b or b==c or a==c:
  if a**2 == b**2 + c**2 or b**2 == a**2 + c**2 or c**2 == a**2 + b**2:
    print ("Tam giác vuông cân")
  else:
     print ("Tam giác cân")
else:
  print ("Tam giác thường")
```

## Bài 8
#### Giải hệ phương trình tuyến tính:
#### ax + by = m
#### cx + dy = n

```python
a,b,c,d,m,n = map (int, input ().split ())
det = a*d - b*c
dx = m*d - b*n
dy = a*n - c*m
if det !=0:
  print (dx/det, dy/det)
else:
  if dx == dy:
    print ("Vô số nghiệm")
  else:
    print ("Vô nghiệm")
```
