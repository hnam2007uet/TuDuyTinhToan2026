## Bài 4
#### Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím
#### Ví dụ: Ta nhập h = 4:
#### *
#### * *
#### * *
#### *******

```python
h = int(input())
for i in range(h - 1):
    print("", end=" ")
print("*")
for i in range(2, h + 1):
    for j in range(1, 2*h):
        if j == h - i + 1 or j == h + i - 1 or i == h:
            print("*", end="")
        else:
            print("", end=" ")
    print()
```

## Bài 8
#### Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b

```python
a, b = map(int, input().split())
n = min(a, b)
m = max(a, b)

res = []

for i in range(1, int(n**0.5) + 1):
    if n % i == 0:
        if m % i == 0:
            res.append(i)
        if i != n // i and m % (n // i) == 0:
            res.append(n // i)

res.sort()
for x in res:
    print(x, end=' ')
```

```python
import math

a, b = map(int, input().split())
g = math.gcd(a, b)

res = []
for i in range(1, int(g**0.5) + 1):
    if g % i == 0:
        res.append(i)
        if i != g // i:
            res.append(g // i)

res.sort ()
print (*res)
```

```python
def is_prime (n):
  if n<2:
    return False
  for i in range (2, int (n**0.5)+1):
    if n%i==0:
      return False
  return True
n = int (input ())
ans = 0
while n!=0:
  if is_prime (n%10):
    ans+=1
  n//=10
print (ans)
```
