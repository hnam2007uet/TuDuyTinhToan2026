## Bài 4
####Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng 6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
####• Các điểm a1, b1, c1 được nhân hệ số 1,
####• Các điểm a2, b2 được nhân hệ số 2,
####• Điểm a3 được nhân hệ số 3.
#### Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10 Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6 đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được phân tách nhau bởi duy nhất 1 dấu cách.Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.


```python
a1, b1, c1, a2, b2, a3 = map (float, input ().split ())
print (f"{(((a1+b1+c1)+(a2+b2)*2+a3*3)/10):.1f}")
```
